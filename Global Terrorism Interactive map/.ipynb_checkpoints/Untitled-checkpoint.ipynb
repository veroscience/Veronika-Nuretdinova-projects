### Calculate top 10 groups for each year

```python
data_short.iyear
```

```python
yearly_data = {}
```

```python
for year in range(1970, 2021):
    yearly_data[year] = data_short[(data_short.iyear==year)&(data_short.gname!="Unknown")].groupby("gname")["nkill"].sum().sort_values(ascending=False)[:10]
```
